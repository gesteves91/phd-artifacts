#### Import the libraries

```python
from pycaret.classification import *
import pandas as pd
```

#### Load unseen data

```python
df = pd.read_csv('../data/unseen/pbeans.csv', index_col=0)
```

#### Remove the unnecessary columns

```python
df = df.drop('LongName', 1)
df = df.drop('Name', 1)
df = df.drop('Parent', 1)
df = df.drop('Component', 1)
df = df.drop('Path', 1)
df = df.drop('Line', 1)
df = df.drop('EndLine', 1)
df = df.drop('Column', 1)
df = df.drop('EndColumn', 1)
```

```python
df.head()
```

#### Load the model (using the GC)

```python
model = load_model('../models/gc')
```

#### Predict the GC code smell using unseen data

```python
predict_model(model, data=df, probability_threshold = 0.3, raw_score=True)
```
